```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)
fig.set_facecolor("skyblue")
ax.grid()
x = np.linspace(-10*np.pi, 10*np.pi, 5000)
y = np.sin(x)/x
ax.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Находит точку пересечения трех прямых
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)
fig.set_facecolor("skyblue")
ax.grid()
x = np.linspace(-2*np.pi, 2*np.pi, 100000)
y1 = x+1
y2 = 2*x+2
y3 = 3*x+3
for_count = 0
values = 0
for el_x in x:
#     print(round(el_x+1, 5))
    if round(el_x+1, 2) == round(el_x*2+2, 2) and round(el_x*2+2, 2) == round(el_x*3+3, 2):
#         print(round(el_x, 2))
        values += round(el_x, 2)
        for_count += 1
if values:
    print(f'x = {round(values/for_count, 2)}, y = {round(values/for_count, 2)+1}')
else:
    print("Not values")
# print("---")
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# Находит точку пересечения двух прямых
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(111)
fig.set_facecolor("skyblue")
ax.grid()
x = np.linspace(-2*np.pi, 2*np.pi, 100000)
y1 = (3*x-12)/4
y2 = (14-x)/7
for_count = 0
values = 0
for el_x in x:
#     print(round(el_x+1, 5))
    if round((3*el_x-12)/4, 2) == round((14-el_x)/7, 2):
#         print(round(el_x, 2))
        values += round(el_x, 2)
        for_count += 1
if values:
    print(f'x = {round(values/for_count, 2)}, y = {round(values/for_count, 2)+1}')
else:
    print("Not values")
# print("---")
ax.plot(x,y1)
ax.plot(x,y2)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
