### 1. Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800]. Найдите ее среднее значение и дисперсию.

'(200' говорит о том, что 200 не входит в значения, но неизвестен шаг, с которым идут очередные значения, тогда возьмем шаг за переменную step:$\newline$
Среднее значение равно $\frac{200+step+800}2 = 500 + \frac{step}2\newline$
Дисперсия равна $\frac{(800-200-step)^2}{12} = \frac{(600-step)^2}{12} = \frac{3600 - 1200step + step^2}{12} = 300 - 100step + \frac{step^2}{12}\newline$
Или:$\newline$
Среднее значение равно $\frac{200+800}2=500\newline$
Дисперсия равна $\frac{(800-200)^2}{12}=\frac{600^2}{12}=\frac{3600}{12}=300\newline$
Тогда зачем там закругленная скобка, указывающая на то, что значение не принадлежит отрезку!?

### 2. О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2. Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5? Если да, найдите ее.

В равномерном распределении $D(X) = \frac{(b-a)^2}{12} => 0.2 = \frac{(b-0.5)^2}{12}\newline$
$(b-0.5)^2 = 0.2*12 = 2.4\newline$
$b-0.5 = \sqrt{2.4} \approx 1.55\newline$
$b = 1.55 + 0.5 = 2.05\newline$
Правая граница получилась 2.05$\newline$
Среднее значение равно $\frac{a+b}{2} = \frac{0.5+2.05}{2}= 1.275$

### 3. Непрерывная случайная величина X распределена нормально и задана плотностью распределения$\newline$f(x) = (1 / (4 * sqrt(2*pi))) * (exp(-((x+2)**2) / 32))$\newline$Найдите:$\newline$а). M(X)$\newline$б). D(X)$\newline$в). std(X) (среднее квадратичное отклонение)

$f(x) = \frac1{\sigma\sqrt{2\pi}}\exp^{-\frac{(x-a)^2}{2\sigma^2}}\newline$
$f(x) = \frac1{4\sqrt{2\pi}}\exp^{-\frac{(x+2)^2}{32}} = \frac1{4\sqrt{2\pi}}\exp^{-\frac{(x-(-2))^2}{2*4^2}}\newline$
$M(X) = -2\newline$
$D(X) = 16\newline$
$std(X) = 4$

### 5. На сколько сигм (средних квадратичных отклонений) отклоняется рост человека, равный 190 см, от математического ожидания роста в популяции, в которой M(X) = 178 см и D(X) = 25 кв.см?

$\sigma = \sqrt{D(X)} = \sqrt{25} = 5\newline$
Отклонение роста от мат.ожидания = $\frac{190-178}{5} = \frac{12}5 = 2.4 (сигм)$
