```python
from matplotlib import pylab as plt
import numpy as np
%matplotlib inline

plt.figure(figsize=(6,6))
plt.axis([-5, 7, -4, 4])

x=np.linspace(-5, 7, 200)
y1=np.sin(x)
y2=-0.25*x+0.5

plt.plot(x,y1, c='b')
plt.plot(x,y2, c='r')
plt.show()
```

```python
import numpy as np
from scipy.optimize import fsolve, broyden2
import math
# x^2-y^2+3*x*y^3-2*x^2*y^2+2*x-3*y-5
# 3*y^3-2*x^2+2*x^3*y-5*x^2*y^2+5

def equations(p):
    x,y = p
    return x^2-y^2+3*x*y^3-2*x^2*y^2+2*x-3*y-5+3*y^3-2*x^2+2*x^3*y-5*x^2*y^2+5

for x in np.arange (-10, 10, 2):
    x, info, ier, mesg =  fsolve(equations, x, full_output=True)
    print (float(x), ier)
```
